In order to use this likelihood:

2. cd desi_dr1_p1d_cobaya_likelihood
3. pip install -e .
4. Then run the notebook

```python
%load_ext autoreload
%autoreload 2

import os
from cobaya.yaml import yaml_load_file
from cobaya.model import get_model
from desi_dr1_p1d_cobaya_likelihood.like_lya_p1d import Like_lya_p1d
```

### Load likelihood

```python
import desi_dr1_p1d_cobaya_likelihood
path_like = desi_dr1_p1d_cobaya_likelihood.__path__
```

```python
info_yaml = os.path.join(path_like[0], "like_lya_p1d.yaml")
packages_path = os.path.dirname(path_like[0])
info = {
    "params":yaml_load_file(info_yaml),
    'likelihood': {'desi_dr1_p1d_cobaya_likelihood.like_lya_p1d': Like_lya_p1d},
    'theory': {'camb': None},
    'packages': packages_path
}
info
```

### Set model

Evaluate the likelihood for a particular value of the H0 parameter, it can be done in the same for other cosmo params

```python
info["params"]["H0"] = 67.77
model = get_model(info)
```

### Evaluate model

```python
model.logposterior({"H0":info["params"]["H0"]})
```
